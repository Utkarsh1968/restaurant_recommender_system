```python
import pandas as pd 
from sklearn.preprocessing import MultiLabelBinarizer,StandardScaler
```

```python
df = pd.read_csv("Dataset .csv")
df
```

```python
df['Cuisines'] = df['Cuisines'].fillna('Unknow')
```

```python
df['Cuisines'].isnull().sum()
```

```python
df['Cuisines'] = df['Cuisines'].str.split(', ')
```

```python
new_df = df[['Cuisines','Aggregate rating','City','Restaurant Name']]
```

```python
new_df
```

```python
df['Cuisines'].value_counts()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
ohe = MultiLabelBinarizer()
```

```python
cuisine_encoded = ohe.fit_transform(new_df['Cuisines'])
cuisine_encoded_df = pd.DataFrame(cuisine_encoded, columns=ohe.classes_)
```

```python
cuisine_encoded_df
```

```python
city_encoded = pd.get_dummies(new_df['City'], prefix='City')
```

```python
combined_df = pd.concat([new_df.drop(['Cuisines', 'City'], axis=1), cuisine_encoded_df, city_encoded], axis=1)
combined_df
```

```python
from sklearn.preprocessing import StandardScaler
mn = StandardScaler()
```

```python
model = mn.fit(combined_df[['Aggregate rating']])
combined_df[['Aggregate rating']]=model.transform(combined_df[['Aggregate rating']])
```

```python
combined_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(combined_df.drop(['Restaurant Name'], axis=1))
```

```python
def get_recommendations(city, cuisines, top_n=5):
    # Filter restaurants based on the provided city and cuisine
    filtered_restaurants = new_df[(new_df['City'] == city) & (new_df['Cuisines'].apply(lambda x: any(cuisine in x for cuisine in cuisines)))]

    if filtered_restaurants.empty:
        return pd.DataFrame(columns=['Restaurant Name'])

    
    filtered_indices = filtered_restaurants.index
    filtered_similarity = similarity_matrix[filtered_indices]

  
    similarity_scores = filtered_similarity.mean(axis=0)
    top_indices = similarity_scores.argsort()[-top_n:][::-1]

    return new_df.iloc[top_indices]['Restaurant Name']
```

```python
user_city = "New Delhi"  # Replace with the desired city
user_cuisine = ['North Indian', 'Chinese']  # Replace with the desired cuisine

# Get recommendations based on user input
recommendations = get_recommendations(user_city, user_cuisine, top_n=5)

print(recommendations)
    
```
